## School Summary

## Per student budget | Average math score | Average reading score

```python

school_summary_df = schooldata_df.copy()
school_summary_df['per_student_budget'] =  (schooldata_df['budget'] / schooldata_df['size']).map("${:,.2f}".format)
reading_score_df = studentdata_df.groupby(by='school_name')['reading_score'].mean(numeric_only=True)
math_score_df = studentdata_df.groupby(by='school_name')['math_score'].mean(numeric_only=True)
school_summary_df = pd.merge(left=school_summary_df, right=reading_score_df, on='school_name', how='left', )
del(reading_score_df)
school_summary_df = pd.merge(left=school_summary_df, right=math_score_df, on='school_name', how='left')
del(math_score_df)
school_summary_df.rename(columns={'reading_score': 'reading_score_avg', 'math_score': 'math_score_avg'}, inplace=True)
school_summary_df['budget'] = school_summary_df['budget'].map("${:,.2f}".format)
school_summary_df['size'] = school_summary_df['size'].map("{:,}".format)
school_summary_df
```

## % passing math (the percentage of students who passed math) | % passing reading (the percentage of students who passed reading)

```python
def agg_pass_mark(x):
  '''
  function to aggregate the percentage passing the subject.
  '''
  pass_mark = 70
  return (x >= pass_mark).mean() * 100
```

```python
pass_math_df = studentdata_df.groupby('school_name', group_keys=False)['math_score'].agg(agg_pass_mark)
school_summary_df = pd.merge(left=school_summary_df, right=pass_math_df, on='school_name', how='left')
pass_reading_df = studentdata_df.groupby('school_name', group_keys=False)['reading_score'].agg(agg_pass_mark)
school_summary_df = pd.merge(left=school_summary_df, right=pass_reading_df, on='school_name', how='left')
school_summary_df.rename(columns={'reading_score': r'% pass_reading', 'math_score': r'% pass_math'}, inplace=True)
school_summary_df
```

## % overall passing (the percentage of students who passed math AND reading

```python
def agg_function_pass_both(x):
  pass_mark = 70
  a = (x['reading_score'] >= pass_mark) & (x['math_score'] >= pass_mark)
  return a.mean() * 100
```

```python
students_passing_df = studentdata_df.groupby('school_name',).apply(agg_function_pass_both)
school_summary_df = pd.merge(left=school_summary_df, right=students_passing_df.to_frame(), on='school_name', how='left')
del(students_passing_df)
school_summary_df.rename(columns={0: r'% overall passing'}, inplace=True)
school_summary_df
```

## Math Scores by Grade

```python
# Use the code provided to separate the data by grade
ninth_graders = studentdata_df[(studentdata_df["grade"] == "9th")]
tenth_graders = studentdata_df[(studentdata_df["grade"] == "10th")]
eleventh_graders = studentdata_df[(studentdata_df["grade"] == "11th")]
twelfth_graders = studentdata_df[(studentdata_df["grade"] == "12th")]

# Group by `school_name` and take the mean of the `math_score` column for each.
ninth_grade_math_scores = ninth_graders.groupby(by=['school_name',])['math_score'].mean(numeric_only=True).rename('9th')
tenth_grader_math_scores = tenth_graders.groupby(by=['school_name',])['math_score'].mean(numeric_only=True).rename('10th')
eleventh_grader_math_scores = eleventh_graders.groupby(by=['school_name',])['math_score'].mean(numeric_only=True).rename('11th')
twelfth_grader_math_scores = twelfth_graders.groupby(by=['school_name',])['math_score'].mean(numeric_only=True).rename('12th')

math_scores_by_grade = pd.concat([ninth_grade_math_scores, tenth_grader_math_scores, eleventh_grader_math_scores, twelfth_grader_math_scores], axis=1)
math_scores_by_grade.index.name = None
math_scores_by_grade
```

## Reading Scores by Grade

```python
# Group by `school_name` and take the mean of the `reading_score` column for each.
ninth_grade_reading_score = ninth_graders.groupby(by=['school_name',])['reading_score'].mean(numeric_only=True).rename('9th')
tenth_grader_reading_score = tenth_graders.groupby(by=['school_name',])['reading_score'].mean(numeric_only=True).rename('10th')
eleventh_grader_reading_score = eleventh_graders.groupby(by=['school_name',])['reading_score'].mean(numeric_only=True).rename('11th')
twelfth_grader_reading_score = twelfth_graders.groupby(by=['school_name',])['reading_score'].mean(numeric_only=True).rename('12th')

reading_scores_by_grade = pd.concat([ninth_grade_reading_score, tenth_grader_reading_score, eleventh_grader_reading_score, twelfth_grader_reading_score], axis=1)
reading_scores_by_grade.index.name = None
reading_scores_by_grade
```

## Scores by School Spending

```python
# Setting up the bins
spending_bins = [0, 585, 630, 645, 680]
labels = ["<$585", "$585-630", "$630-645", "$645-680"]
```

```python
school_summary_df['spending_bins'] = pd.cut(x= school_summary_df['per_student_budget'].str.replace(pat=r'$', repl='',regex=False).astype(float),
                                            bins=spending_bins,
                                            labels=labels)
# school_summary_df
```

```python
spending_summary = school_summary_df.sort_values(by='per_student_budget')
spending_summary
```

## Scores by School Size

```python
size_bins = [0, 1000, 2000, 5000]
size_labels = ["Small (<1000)", "Medium (1000-2000)", "Large (2000-5000)"]
```

```python
school_summary_df['size_bins'] = pd.cut(x=school_summary_df['size'].str.replace(pat=r',', repl='',regex=False).astype(int),
                                        bins=size_bins,
                                        labels=size_labels)
```

```python
size_summary = school_summary_df.sort_values(by='size')
size_summary
```

## Scores by School Type

```python
# By school type
average_math_score_by_type = school_summary_df.groupby(["type"])["math_score_avg"].mean()
average_reading_score_by_type = school_summary_df.groupby(["type"])["reading_score_avg"].mean()
average_percent_passing_math_by_type = school_summary_df.groupby(["type"])["% pass_math"].mean()
average_percent_passing_reading_by_type = school_summary_df.groupby(["type"])["% pass_reading"].mean()
average_percent_overall_passing_by_type = school_summary_df.groupby(["type"])["% overall passing"].mean()

# Assemble
school_type_summary = pd.DataFrame(
    {
        "Average Math Score" : average_math_score_by_type,
        "Average Reading Score": average_reading_score_by_type,
        "% Passing Math": average_percent_passing_math_by_type,
        "% Passing Reading": average_percent_passing_reading_by_type,
        "% Overall Passing": average_percent_overall_passing_by_type
    }
)

school_type_summary
```
